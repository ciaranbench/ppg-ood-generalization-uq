```python
import numpy as np
import pickle
from sklearn.neighbors import NearestNeighbors
```

```python
def compile_latents(filename):
    latents = np.load(filename,allow_pickle=True)
    all_latents = np.concatenate(
        [np.squeeze(item) for sublist in latents for item in sublist],
        axis=0
    )
    print(all_latents.shape)
    return all_latents

def l2_norm(x):
    return x / np.linalg.norm(x, axis=1, keepdims=True)

def knn_calc(latents_ref,test_latents,k=5):

    nn = NearestNeighbors(n_neighbors=k, metric='euclidean')
    nn.fit(l2_norm(latents_ref))  # (25000, 2048)
    test_latents = l2_norm(test_latents)
    distances, _ = nn.kneighbors(test_latents, n_neighbors=k)

    
    print(f"k-NN (k=5) Mean - Holdout: {distances[:, k-1].mean():.2f}")

    
```

```python
latents_ref = compile_latents('test_latents.pkl')[:20000]
#test_latents= compile_latents('test_latents.pkl')[20000:40000]
test_latents = compile_latents('test_latents.pkl')[:20000]
```

```python
knn_calc(latents_ref,test_latents)
```
